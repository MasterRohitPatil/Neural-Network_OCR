```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_prediction(train_data=x_train,train_labels=y_train,test_data=x_test,test_labels=y_test,predictions=y_preds):
  #Plot training data, test data and compares predictions.
  plt.figure(figsize=(10,7))
   # plot training data in blue
  plt.scatter(train_data,train_labels,c="b",label="Training data")
  #plot test data in green
  plt.scatter(test_data,test_labels,c="g",label="Testing data")
  #plot predictions in red (predictions were made on test data )
  plt.scatter(test_data,predictions,c="r",label="Predictions")
  #show legend
  plt.legend()
```

```python
x = np.arange(-100,100,4)
x
```

```python
#same result as above with plus 10
y= x+10
y
```

```python
#split data into train and  test sets
x_train =x[:40] #last 40 examples (80% of data)
y_train = y[:40]

x_test = x[40:]#last 10 examples (20% of data )
y_test = y[40:]
len(x_train),len(x_test)
```

```python
plt.figure(figsize=(10,7))
#plot training data in blue
plt.scatter(x_train,y_train,c="b",label="Training data")
#plot test data in green
plt.scatter(x_test,y_test,c="g",label="Testing data")
# show the legend
plt.legend()
```

```python
#set random seed
import tensorflow as tf
tf.random.set_seed(42)
#create a model (same as above)
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1, input_shape=[1])
])
#compile model (same as above)
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
model.fit(x_train,y_train,epochs=100) #verbose controls how much gets output
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model,show_shapes=True)
```

```python
y_preds = model.predict(x_test) # make predictions
y_preds # view predictions
```

```python
plot_prediction(train_data=x_train,train_labels=y_train,test_data=x_test,test_labels=y_test,predictions=y_preds)
```
